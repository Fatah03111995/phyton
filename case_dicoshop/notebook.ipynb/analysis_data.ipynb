```python
# DUPLICATED DATA
customers_df.drop_duplicates(inplace = True)
print("jumlah duplikasi", customers_df.duplicated().sum())
```

```python
#MISSING VALUE
customers_df[customers_df.gender.isnull()]
```

```python
customers_df.fillna("Prefer not to say", inplace=True) #METODE IMPUTATION
```

```python
customers_df.isna().sum()
```

```python
#MEMBERSIHKAN DATA ORDERS
datetime_columns = ["order_date","delivery_date"]
for column in datetime_columns:
    orders_df[column] = pd.to_datetime(orders_df[column])

orders_df.info()
```

```python
print("duplicate : ", orders_df.duplicated().sum())
orders_df.drop_duplicates(inplace=True)
```

```python
delivery_time = orders_df['delivery_date'] - orders_df['order_date'] #hanya berisi index dan selisih hari
delivery_time.info()
delivery_time
```

```python
delivery_time = delivery_time.apply(lambda x: x.total_seconds())
delivery_time
```

```python
orders_df['delivery_time'] = round(delivery_time/86400)
orders_df.describe(include="all")
```

```python
customers_active = orders_df.customer_id.tolist()
customers_df['status'] = customers_df['customer_id'].apply(lambda x : 'Active' if x in customers_active else 'Non Active')
customers_df.sample(5)
```

```python
customers_df.groupby(by='status').customer_id.count()
```

```python
customers_df.groupby(by='status').customer_id.nunique()
```

```python
merge_df = pd.merge(
    left = orders_df,
    right = customers_df,
    how = 'left',
    left_on = 'customer_id',
    right_on = 'customer_id'
)

merge_df.head()
```

```python
merge_df.groupby(by='city').order_id.nunique()
```

```python
merge_df.groupby(by='city').order_id.nunique().sort_values(ascending=False)
```
