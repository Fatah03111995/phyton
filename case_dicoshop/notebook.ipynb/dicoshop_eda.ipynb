```python
import pandas as pd

customers_df = pd.read_csv("../customers.csv", sep=',')
products_df = pd.read_csv("../products.csv",sep=',')
orders_df = pd.read_csv("../orders.csv", sep=',')
sales_df = pd.read_csv("../sales.csv", sep=',')

customers_df.info()
```

```python
customers_df.drop_duplicates(inplace=True) #menghilangkan duplikasi
customers_df.fillna('Prefer not to say', inplace = True) #mengisi nilai na dengan prefer not to say)
customers_df[customers_df.age == customers_df['age'].max()] #gk mungkin ada umur 700
customers_df.age.replace(customers_df.age.max(), 70, inplace=True)
```
