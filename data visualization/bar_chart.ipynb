```python
import matplotlib.pyplot as plt

cities = ('Bogor', 'Bandung', 'Jakarta', 'Semarang', 'Yogyakarta', 
          'Surakarta','Surabaya', 'Medan', 'Makassar')
 
populations = (45076704, 11626410, 212162757, 19109629, 50819826, 17579085,
               3481, 287750, 785409)

plt.bar(x=cities, height=populations)
plt.show()
```

```python
plt.bar(x=cities, height=populations)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.barh(y=cities, width=populations)
plt.show()
```

```python
import pandas as pd

data_frame = pd.DataFrame({
    'Cities': cities,
    'Population': populations
})

data_frame
```

```python
data_frame.groupby(by='Cities').Population.sum().sort_values(ascending=False).reset_index()
```

```python
data_frame.sort_values(by='Population', inplace=True)
data_frame
```

```python
plt.barh(y=data_frame['Cities'], width=data_frame['Population'])
plt.show()
```

```python
plt.barh(y=data_frame['Cities'], width=data_frame['Population'])
plt.xlabel('Population (Millions)')
plt.ylabel('Cities')
plt.title('Population Ciies in Indonesia')
plt.show()
```

```python
import seaborn as sns

data_frame.sort_values(by='Population', inplace=True, ascending=False)

sns.barplot(y=data_frame['Cities'], x=data_frame['Population'], orient='h', color='red')
plt.title('Popoulation Cities in Indonesia')
plt.xlabel('Population (Millions)')
plt.ylabel('Cities')
plt.show()
```

```python
data_frame.sort_values(by='Population', inplace=True, ascending=False)

sns.barplot(x=data_frame['Cities'], y=data_frame['Population'], orient='v', color='blue')
plt.title('Population Cities in Indonesia')
plt.xlabel('Cities')
plt.xticks(rotation=45)
plt.ylabel('Population (Million)')
plt.show()
```
