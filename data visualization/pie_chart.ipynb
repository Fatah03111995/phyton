```python
import matplotlib.pyplot as plt

flavors = ('Chocolate', 'Vanilla', 'Macha', 'Others')
votes = (50, 20, 30, 10)
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D')
explode = (0.05, 0, 0, 0)

plt.pie(
    x=votes,
    labels=flavors,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode
)
plt.show()
```

```python
# Parameter x=votes berisi nilai yang ingin ditampilkan dalam bentuk pie chart.
# labels=flavors menentukan label dari setiap segmen sesuai dengan data flavors.
# autopct='%1.1f%%' digunakan untuk menampilkan persentase tiap segmen, dengan satu angka desimal.
# colors=colors menetapkan warna untuk setiap bagian pie chart.
# explode=explode menetapkan bagian yang ingin sedikit dikeluarkan dari lingkaran pusat.
```

```python
plt.pie(
    x=votes,
    labels=flavors,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode,
    wedgeprops={'width':0.3}
)
plt.show()
```
